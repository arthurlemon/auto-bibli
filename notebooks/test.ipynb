```python
import sys
import requests
from pprint import pprint
sys.path.append("../")
from src.utils import search_book, check_availability

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()),
                          options=options)

# Step 1: Open the advanced search page URL
BASE_URL = "https://nelligandecouverte.ville.montreal.qc.ca/iii/encore"
search_url = f"{BASE_URL}/home?lang=frc&suite=cobalt&advancedSearch=true&searchString="
```

```python
driver.implicitly_wait(2)

driver.get(search_url)

# Step 2: Fill in the "Titre et collection" field
title = "the wake"
title_select = Select(driver.find_element(By.ID, 'searchType_0'))
title_select.select_by_value('t:')  # "Titre et collection" option
title_input = driver.find_element(By.ID, 'searchTerm_0')
title_input.send_keys(title)

# Step 3: Fill in the "Collection" field
collection_select = Select(driver.find_element(By.ID, 'limitValue_1'))
collection_select.select_by_value('55')  # "Adultes" option

# Step 4: Fill in the "Localisation" field
location_select = Select(driver.find_element(By.ID, 'limitValue_2'))
location_select.select_by_value('x33a')  # "LE PRÃ‰VOST - Adultes" option

# Step 5: Fill in the "Format" field
format_select = Select(driver.find_element(By.ID, 'limitValue_0'))
format_select.select_by_value('-')  # "Livre" option

# Step 6: Fill in the "Langue" field
language_select = Select(driver.find_element(By.ID, 'limitValue_3'))
language_select.select_by_value('eng')  # Francais & "Anglais"

# Step 7: Submit the form
search_button = driver.find_element(By.ID, 'searchSubmit')
search_button.click()

# Step 8: Get the resulting URL
result_url = driver.current_url
print("Search Result URL:", result_url)
```
